## Dataset Preprocess

```python
import os
import random
import shutil
```

```python
datasetPath = r'C:\Users\johnp\Desktop\Dataset\cassava_all_dataset\Classification'
outputPath = r'outputDataset'
```

```python
val_size = 60
train_size = 240
```

```python
# Create the output directory for the class
os.makedirs(os.path.join(outputPath, "val"), exist_ok=True)
os.makedirs(os.path.join(outputPath, "train"), exist_ok=True)
```

```python
random.seed(42)

for class_name in os.listdir(datasetPath):
    
    class_path = os.path.join(datasetPath, class_name)
    val_dir = os.path.join(outputPath, "val", class_name)
    train_dir = os.path.join(outputPath, "train", class_name)
    
    os.makedirs(val_dir, exist_ok=True)
    os.makedirs(train_dir, exist_ok=True)

    
    all_images = [os.path.join(class_path, img) for img in os.listdir(class_path)]
    random.shuffle(all_images)

    
    val_files = all_images[:val_size]
    remaining_files = all_images[val_size:]

    
    if len(remaining_files) < train_size:
        needed = train_size - len(remaining_files)
        
        extra = random.choices(all_images, k=needed) 
        train_files = remaining_files + extra
    else:
        train_files = remaining_files[:train_size]

    
    for img in val_files:
        shutil.copy(img, val_dir)
        
    for img in train_files:
        shutil.copy(img, train_dir)
```

### Dataset train val split

```python
all_images = []

for folder in os.listdir(datasetPath):
    folderPath = os.path.join(datasetPath, folder)
    
    for image in os.listdir(folderPath):
        imagePath = os.path.join(folderPath, image)
        
        all_images.append(imagePath)

    random.shuffle(all_images)
    split_index = int(0.8 * len(all_images))
    train_images = all_images[:split_index]    
    val_images = all_images[split_index:]
    for imagePath in train_images:
        dest_folder = os.path.join(outputPath, 'train', folder)
        os.makedirs(dest_folder, exist_ok=True)
        shutil.copy(imagePath, dest_folder)

    for imagePath in val_images:
        dest_folder = os.path.join(outputPath, 'val', folder)
        os.makedirs(dest_folder, exist_ok=True)
        shutil.copy(imagePath, dest_folder)
```

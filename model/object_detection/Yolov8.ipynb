```python
from ultralytics import YOLO
```

```python
EPOCHS = 40
BATCH_SIZE = 16
Data_Dir = r'C:\Users\e6va6je238\Desktop\Cassava_Leaf_Datasets\Object_Detection/data.yaml'
Image_Size = 640
```

```python
model = YOLO('yolov8m.pt')
```

#### Phase 1

```python
model.train(
    data=Data_Dir,
    epochs=EPOCHS,
    imgsz=Image_Size,
    batch=BATCH_SIZE,          
    lr0=0.001,
    optimizer="SGD",
    weight_decay=0.0005,
    patience=20,
    freeze=10,
    project="YOLOv8_Cassava_Leaf_Detection",       
    name="leaf_stage_1"
)
```

#### Phase 2

```python
model = YOLO("YOLOv8_Cassava_Leaf_Detection/leaf_stage_1/weights/best.pt")

model.train(
    data=Data_Dir, 
    epochs=40,
    imgsz=640,
    lr0=0.0001,
    batch=BATCH_SIZE,
    project="YOLOv8_Cassava_Leaf_Detection",
    name="leaf_stage_2"
)
```
